```python
with open('Add_employee.csv', 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) 


    for idx, row in enumerate(csv_reader, start=1):
        if idx <= 23: 
            EmployeeID_input = driver.find_element("id", "employeeId")
            EmployeeID_input.clear()

            EmployeeID = row[2]
            FirstName = row[3]
            LastName = row[5]
        
            FirstName_input = driver.find_element("id", "firstName")
            FirstName_input.send_keys(FirstName)

            LastName_input = driver.find_element("id", "lastName")
            LastName_input.send_keys(LastName)

            EmployeeID_input.send_keys(EmployeeID)

            Save_button = driver.find_element("id", "btnSave")
            Save_button.click()
            temp = True
            temp1 = True
            try:
                wait = WebDriverWait(driver, 1)
                error_message_element = wait.until(EC.presence_of_element_located((By.XPATH, "//span[@class='validation-error']")))
                error_message = error_message_element.text
                print("Thông báo lỗi:", error_message)
                temp1 = False
            except TimeoutException:
                temp1 = True

            try:
                wait = WebDriverWait(driver, 1)
                error_message_element = wait.until(EC.presence_of_element_located((By.XPATH, "//div[@class='message warning fadable']")))
                error_message = error_message_element.text
                print("Thông báo lỗi:", error_message)
                temp = False
            except TimeoutException:
                temp = True

            if (temp == False or temp1 == False):       
                with open('Add_employee.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                rows[idx][7] = "INVALID" 

                with open('Add_employee.csv', 'w', newline='') as file:
                    csv_writer = csv.writer(file)
                    csv_writer.writerows(rows)
                

                #Result
                with open('Add_employee.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                if rows[idx][6] == rows[idx][7]:  
                    rows[idx][8] = "PASS" 

                    with open('Add_employee.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)
                
                else: 
                    rows[idx][8] = "FAIL"
                    with open('Add_employee.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)
                    Rec_button = driver.find_element("id", "menu_pim_viewPimModule")
                    Rec_button.click()

                    Add_button = driver.find_element("id", "menu_pim_addEmployee")
                    Add_button.click()
            else:
                print("Bạn thêm thành công")

                with open('Add_employee.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                rows[idx][7] = "VALID"  

                with open('Add_employee.csv', 'w', newline='') as file:
                    csv_writer = csv.writer(file)
                    csv_writer.writerows(rows)

                #Result
                with open('Add_employee.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                if rows[idx][6] == rows[idx][7]:  
                    rows[idx][8] = "PASS"  

                    with open('Add_employee.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)
                else: 
                    rows[idx][8] = "FAIL"
                    with open('Add_employee.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)

                Rec_button = driver.find_element("id", "menu_pim_viewPimModule")
                Rec_button.click()

                Add_button = driver.find_element("id", "menu_pim_addEmployee")
                Add_button.click()
```
