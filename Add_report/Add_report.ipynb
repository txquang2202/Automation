```python
with open('Add_report.csv', 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Bỏ qua dòng tiêu đề


    for idx, row in enumerate(csv_reader, start=1):
        if idx <= 4: 
            report_input = driver.find_element("id", "report_report_name")
            report_input.clear()

            report = row[1]
            Criteria = row[2]
            Field_Groups = row[3]
            Fields = row[4]     

            report_input.send_keys(report)

            Selection_Criteria = driver.find_element("id", "report_criteria_list")
            Dropdown_Criteria = Select(Selection_Criteria)
            Dropdown_Criteria.select_by_visible_text(Criteria)
            if(idx != 5 or idx !=6 or idx != 8):
                adding_button = driver.find_element("id", "btnAddConstraint")
                adding_button.click()

            Display_Fields = driver.find_element("id", "report_display_field_list")
            Dropdown_Fields = Select(Display_Fields)
            Dropdown_Fields.select_by_visible_text(Fields)
            if(idx < 6 ):
                adding_button = driver.find_element("id", "btnAddDisplayField")
                adding_button.click()
                
            Display_Field_Groups = driver.find_element("id", "report_display_groups")
            Dropdown_Field_Group = Select(Display_Field_Groups)
            Dropdown_Field_Group.select_by_visible_text(Field_Groups)
            if(idx != 5 or idx != 7 or idx != 8):
                adding_button = driver.find_element("id", "btnAddDisplayGroup")
                adding_button.click()

            Save_button = driver.find_element("id", "btnSave")
            Save_button.click()
            temp = True
            try:
                wait = WebDriverWait(driver, 1)
                error_message_element = wait.until(EC.presence_of_element_located((By.XPATH, "//span[@class='validation-error']")))
                error_message = error_message_element.text
                print("Thông báo lỗi:", error_message)
                temp = False
            except TimeoutException:
                temp = True

            if (temp == False):       
                with open('Add_report.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                rows[idx][6] = "INVALID" 

                with open('Add_report.csv', 'w', newline='') as file:
                    csv_writer = csv.writer(file)
                    csv_writer.writerows(rows)
                

                #Result
                with open('Add_report.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                if rows[idx][5] == rows[idx][6]:  
                    rows[idx][7] = "PASS" 

                    with open('Add_report.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)
                
                else: 
                    rows[idx][7] = "FAIL"
                    with open('Add_report.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)
                    Rec_button = driver.find_element("id", "menu_pim_viewPimModule")
                    Rec_button.click()

                    report_button = driver.find_element("id", "menu_core_viewDefinedPredefinedReports")
                    report_button.click()

                    Add_button = driver.find_element("id", "btnAdd")
                    Add_button.click()

            else:
                print("Bạn thêm thành công")

                with open('Add_report.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                rows[idx][6] = "VALID"  

                with open('Add_report.csv', 'w', newline='') as file:
                    csv_writer = csv.writer(file)
                    csv_writer.writerows(rows)

                #Result
                with open('Add_report.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                if rows[idx][5] == rows[idx][6]:  
                    rows[idx][7] = "PASS"  

                    with open('Add_report.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)
                else: 
                    rows[idx][7] = "FAIL"
                    with open('Add_report.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)

                Rec_button = driver.find_element("id", "menu_pim_viewPimModule")
                Rec_button.click()

                report_button = driver.find_element("id", "menu_core_viewDefinedPredefinedReports")
                report_button.click()

                Add_button = driver.find_element("id", "btnAdd")
                Add_button.click()
```
