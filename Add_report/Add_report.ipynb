```python
with open('Add_report.csv', 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Bỏ qua dòng tiêu đề


    for idx, row in enumerate(csv_reader, start=1):
        if idx <= 12: 
            #report 
            report_input = driver.find_element("id", "report_report_name")
            report_input.clear()

            report = row[2]
            Criteria = row[3]
            Criteria_include = row[4]
            Fields = row[5]     

            report_input.send_keys(report)

            #citeria
            if Criteria:
                Dropdown_Criteria_driver = driver.find_element("id", "report_criteria_list")

                Dropdown_Criteria = Select(Dropdown_Criteria_driver)
                Dropdown_Criteria.select_by_visible_text(Criteria)
                if(idx != 8):
                    adding_button = driver.find_element("id", "btnAddConstraint")
                    adding_button.click()

            #citeria include
            if Criteria_include:
                Dropdown_Criteria_Include_driver = driver.find_element(By.ID, "report_include_comparision")
                Dropdown_Criteria_Include = Select(Dropdown_Criteria_Include_driver)
                Dropdown_Criteria_Include.select_by_visible_text(Criteria_include)
            #field
            if Fields:    
                Display_Field_Groups_driver = driver.find_element("id", "report_display_groups")
                Dropdown_Field_Group = Select(Display_Field_Groups_driver)
                Dropdown_Field_Group.select_by_visible_text(Fields)
                if(idx != 12):
                    adding_button = driver.find_element("id", "btnAddDisplayGroup")
                    adding_button.click()

            Save_button = driver.find_element("id", "btnSave")
            Save_button.click()
            temp = True
            try:
                wait = WebDriverWait(driver, 1)
                error_message_element = wait.until(EC.presence_of_element_located((By.XPATH, "//span[@class='validation-error']")))
                error_message = error_message_element.text
                print("Thông báo lỗi:", error_message)
                temp = False
            except TimeoutException:
                temp = True

            if (temp == False):       
                with open('Add_report.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                rows[idx][7] = "INVALID" 

                with open('Add_report.csv', 'w', newline='') as file:
                    csv_writer = csv.writer(file)
                    csv_writer.writerows(rows)
                

                #Result
                with open('Add_report.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                if rows[idx][6] == rows[idx][7]:  
                    rows[idx][8] = "PASS" 

                    with open('Add_report.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)
                    
                    Rec_button = driver.find_element("id", "menu_pim_viewPimModule")
                    Rec_button.click()

                    report_button = driver.find_element("id", "menu_core_viewDefinedPredefinedReports")
                    report_button.click()

                    Add_button = driver.find_element("id", "btnAdd")
                    Add_button.click()
                
                else: 
                    rows[idx][8] = "FAIL"
                    with open('Add_report.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)
                    Rec_button = driver.find_element("id", "menu_pim_viewPimModule")
                    Rec_button.click()

                    report_button = driver.find_element("id", "menu_core_viewDefinedPredefinedReports")
                    report_button.click()

                    Add_button = driver.find_element("id", "btnAdd")
                    Add_button.click()

            else:
                print("Bạn thêm thành công")

                with open('Add_report.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                rows[idx][7] = "VALID"  

                with open('Add_report.csv', 'w', newline='') as file:
                    csv_writer = csv.writer(file)
                    csv_writer.writerows(rows)

                #Result
                with open('Add_report.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                if rows[idx][6] == rows[idx][7]:  
                    rows[idx][8] = "PASS"  

                    with open('Add_report.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)
                else: 
                    rows[idx][8] = "FAIL"
                    with open('Add_report.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)

                Rec_button = driver.find_element("id", "menu_pim_viewPimModule")
                Rec_button.click()

                report_button = driver.find_element("id", "menu_core_viewDefinedPredefinedReports")
                report_button.click()

                Add_button = driver.find_element("id", "btnAdd")
                Add_button.click()
```
