```python
with open('Search_Employee.csv', 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Bỏ qua dòng tiêu đề


    for idx, row in enumerate(csv_reader, start=1):
        if idx <= 15: 

            Employee_Name = row[2]
            Id = row[3]
            Employee_Status = row[4] 
            job_title = row[5]

            #name
            Employee_Name_input = driver.find_element("id", "empsearch_employee_name_empName")
            Employee_Name_input.send_keys(Employee_Name)

            #id
            Id_input = driver.find_element("id", "empsearch_id")
            Id_input.send_keys(Id)

            #emp status
            if Employee_Status:
                Employee_Status_driver = driver.find_element("id", "empsearch_employee_status")
                Dropdown_status = Select(Employee_Status_driver)
                Dropdown_status.select_by_visible_text(Employee_Status)

            #job title    
            if job_title:
                job_title_driver = driver.find_element("id", "empsearch_job_title")
                Dropdown_job = Select(job_title_driver)
                Dropdown_job.select_by_visible_text(job_title)


            Search_button = driver.find_element("id", "searchBtn")
            Search_button.click()
            temp = True
            try:
                wait = WebDriverWait(driver, 1)
                error_message_element = wait.until(EC.presence_of_element_located((By.XPATH, "//span[@class='validation-error']")))
                error_message = error_message_element.text
                print("Thông báo lỗi:", error_message)
                temp = False
            except TimeoutException:
                temp = True

            if (temp == False):       
                with open('Search_Employee.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                rows[idx][7] = "INVALID" 

                with open('Search_Employee.csv', 'w', newline='') as file:
                    csv_writer = csv.writer(file)
                    csv_writer.writerows(rows)
                

                #Result
                with open('Search_Employee.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                if rows[idx][6] == rows[idx][7]:  
                    rows[idx][8] = "PASS" 

                    with open('Search_Employee.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)
                
                else: 
                    rows[idx][8] = "FAIL"
                    with open('Search_Employee.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)
                    
                    Rec_button = driver.find_element("id", "menu_pim_viewPimModule")
                    Rec_button.click()
            else:
                print("Bạn tìm thành công")

                with open('Search_Employee.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                rows[idx][7] = "VALID"  

                with open('Search_Employee.csv', 'w', newline='') as file:
                    csv_writer = csv.writer(file)
                    csv_writer.writerows(rows)

                #Result
                with open('Search_Employee.csv', 'r') as file:
                    csv_reader = csv.reader(file)
                    rows = list(csv_reader)

                if rows[idx][6] == rows[idx][7]:  
                    rows[idx][8] = "PASS"  

                    with open('Search_Employee.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)
                else:  
                    rows[idx][8] = "FAIL"
                    with open('Search_Employee.csv', 'w', newline='') as file:
                        csv_writer = csv.writer(file)
                        csv_writer.writerows(rows)
                   
                    Rec_button = driver.find_element("id", "menu_pim_viewPimModule")
                    Rec_button.click()
```
